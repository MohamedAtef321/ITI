```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
from datetime import datetime

# Initialize black image
img = np.zeros((500,500,3), np.uint8)

# Define parameters
center = (int(img.shape[0]/2), int(img.shape[1]/2))
l1 = 0.6     #
l2 = 0.65    # Define how much percentage line length of the whole clock radius
l3 = 0.8     #
r = 140   # clock radius
t = 1     # clock thickness
r_smallcir = 1   # radius of big circles (Numbers)
r_bigcir = 2     # radius of small circles (between numbers)
text_scale = 1   # Date/Time Text Scale
text_thick = 2   # Date/Time Text thickness

# initialize big red circle (clock)
cv2.circle(img, center, r, (0,0,255), t)

# Drawing little yellow and gray circles
for ang in range(0,360,6):
    # define dx, dy to add them to center of clock
    delta_x_clock = r*math.cos(math.radians(ang))
    delta_y_clock = r*math.sin(math.radians(ang))

    # if pass 30 degree we should draw big circle
    if ang%30 == 0:
        # define dx_bigcir, dy_bigcir to subtract them from point
        delta_x_bigcir = r_bigcir*math.cos(math.radians(ang))
        delta_y_bigcir = r_bigcir*math.sin(math.radians(ang))

        x = int(round(center[0] + delta_x_clock - delta_x_bigcir))
        y = int(round(center[1] + delta_y_clock - delta_y_bigcir))

        # draw yellow big circle
        cv2.circle(img, (x,y), r_bigcir, (0,255,255),-1)

    else : # else then we should draw gray small circle
        # define dx_smallcir, dy_smallcir to subtract them from point
        delta_x_smallcir = r_smallcir*math.cos(math.radians(ang))
        delta_y_smallcir = r_smallcir*math.sin(math.radians(ang))    

        x = int(round(center[0] + delta_x_clock - delta_x_smallcir))
        y = int(round(center[1] + delta_y_clock - delta_y_smallcir))

        # draw gray small circle
        cv2.circle(img, (x,y), r_smallcir, (150,150,150),-1)

# saving the clock background to not create it again
background = img.copy()

while True:
    # Get Date and Time
    now = datetime.now()
    date = now.strftime('%Y-%m-%d')
    hour = now.hour
    minute = now.minute
    second = now.second
    time = now.strftime('%H:%M:%S')

    # get image from background
    img = background.copy()

    # Adding Date and Time to image
    cv2.putText(img, str('Date : '+date), (10,40), cv2.FONT_HERSHEY_SIMPLEX,
                text_scale, (255,255,0), text_thick)
    cv2.putText(img, str('Time : '+time), (10,80), cv2.FONT_HERSHEY_SIMPLEX,
                text_scale, (255,255,0), text_thick)

    # Define angles of Clockwise
    angle_1 = hour*30 - 90 + (minute/60)*30
    angle_2 = minute*6 - 90 + (second/60)*6
    angle_3 = second*6 - 90

    # Define end point of hour line
    xl1 = int(round(center[0] + l1* r*math.cos(math.radians(angle_1))))
    yl1 = int(round(center[1] + l1* r*math.sin(math.radians(angle_1))))
    # Define end point of minute line
    xl2 = int(round(center[0] + l2* r*math.cos(math.radians(angle_2))))
    yl2 = int(round(center[1] + l2* r*math.sin(math.radians(angle_2))))
    # Define end point of second line
    xl3 = int(round(center[0] + l3* r*math.cos(math.radians(angle_3))))
    yl3 = int(round(center[1] + l3* r*math.sin(math.radians(angle_3))))

    # Draw Clockwise lines
    cv2.line(img, center, (xl1,yl1), (150,150,0), 2)
    cv2.line(img, center, (xl2,yl2), (50,200,0), 2)
    cv2.line(img, center, (xl3,yl3), (0,255,255), t)

    # Draw gray small-centered circle
    cv2.circle(img, center, 4, (100,100,100), -1)

    # Display Image
    cv2.imshow('Image', img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break    

cv2.destroyAllWindows()
```
