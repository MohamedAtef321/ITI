# Circle 
- Center
- Raduis
- Color
- Thickness

```python
img = np.zeros((500, 500, 3), np.uint8)
cv2.circle(img, (250,250), 100, (255,0,0), -1 )
plt.imshow(img)
```

```python
img = np.zeros((500, 500, 3), np.uint8)
cv2.circle(img, (250,250), 100, (255,0,0), 10 )
plt.imshow(img)
```

```python
import math 
img = np.zeros((500, 500, 3), np.uint8)

cv2.circle(img, (250,250), 100, (255,0,0), 5 )
cv2.line(img, (150,250),(350,250),(255,0,0),5)
cv2.line(img, (250,150),(250,350),(255,0,0),5)

x = int(250 + 100*math.cos(math.radians(45)))
y = int(250 + 100*math.sin(math.radians(45)))
cv2.line(img,(250,250), (x,y), (0,255,0), 5)

x = int(250 + 100*math.cos(math.radians(135)))
y = int(250 + 100*math.sin(math.radians(135)))
cv2.line(img,(250,250), (x,y), (0,255,0), 5)

x = int(250 + 100*math.cos(math.radians(225)))
y = int(250 + 100*math.sin(math.radians(225)))
cv2.line(img,(250,250), (x,y), (0,255,0), 5)

x = int(250 + 100*math.cos(math.radians(315)))
y = int(250 + 100*math.sin(math.radians(315)))
cv2.line(img,(250,250), (x,y), (0,255,0), 5)

plt.figure(figsize=(8,8))
plt.imshow(img)
```

# polygon
- array of points of type int32
- isClosed
- Color
- Thickness

```python
img = np.zeros((500,500, 3), np.uint8)
pts = np.array([
    [100,250], [250,300], [250,250], [50,60]
    ],np.int32)
cv2.polylines(img,[pts], True, (225,225,0), 3)

plt.imshow(img);
```

```python
img = np.zeros((400,800, 3), np.uint8)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, 'Tekomoro', (10,250), font, 5, (225,225,0), 3  )

plt.figure(figsize=(8,8))
plt.imshow(img);
```
