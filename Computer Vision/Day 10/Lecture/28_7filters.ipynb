```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
src = np.array(
    [
        [7,2,3,3,8],
        [4,5,3,8,4],
        [3,3,2,8,4],
        [2,8,7,2,7],
        [5,4,4,5,4]
    ]
    ,dtype= np.float32
)

kernel = np.array(
    [
        [1,0,-1],
        [1,0,-1],
        [1,0,-1]
    ]
    ,dtype= np.float32
)

out = cv2.filter2D(src, -1, kernel,borderType= cv2.BORDER_ISOLATED )
print(out)

f,(ax0, ax1, ax2) = plt.subplots(1,3,figsize= (10,10))
ax0.imshow(src,cmap='gray')
ax0.set_title('source')
ax1.imshow(kernel,cmap='gray')
ax1.set_title('kernel')
ax2.imshow(out,cmap='gray')
ax2.set_title('output');
```

```python
img = cv2.imread('Files/001.jpg')


sobel_y= np.array(
    [
        [-1,-1,-1],
        [0,0,0],
        [1,1,1]
    ]
,np.float32
)

sobel_x= np.array(
    [
        [1,0,-1],
        [1,0,-1],
        [1,0,-1]
    ]
,np.float32
)
sobel  = sobel_y + sobel_x
out = cv2.filter2D(img,-1,sobel,borderType=cv2.BORDER_ISOLATED)
cv2.imshow('sobel',out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('Files/001.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
kernel1 = np.ones((5,5),np.float32)/25
dst1 = cv2.filter2D(img,-1,kernel1,borderType=cv2.BORDER_ISOLATED)

kernel2 = np.ones((25,25),np.float32)/625
dst2 = cv2.filter2D(img,-1,kernel2,borderType=cv2.BORDER_ISOLATED)

f,(ax0, ax1, ax2) = plt.subplots(1,3,figsize= (30,50))
ax0.imshow(img)
ax1.imshow(dst1)
ax2.imshow(dst2);
```

```python
blur = cv2.blur(img,(20,20))
plt.imshow(blur);
```

```python
def invert(img):
    inv = cv2.bitwise_not(img)
    return inv
```

```python
out = invert(img)
cv2.imshow('invert',out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gaussian = cv2.GaussianBlur(img, (9,9), 0)

f, (ax0, ax1) = plt.subplots(1, 2, figsize=(25,10))
ax0.imshow(img), ax0.set_title('Original')
ax1.imshow(gaussian), ax1.set_title('Gaussian');
```
