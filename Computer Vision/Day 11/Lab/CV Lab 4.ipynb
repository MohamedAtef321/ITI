```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Q1- load the following image and make red rectangle around face and write below
# rectangle “Cat” with yellow color
img = plt.imread('original.png')
cv2.rectangle(img, (780,120), (1500,820), (255,0,0), 10)
cv2.putText(img, 'Cat', (780,950), cv2.FONT_HERSHEY_SIMPLEX, 5, (255,255,0), 10)
plt.imshow(img)
```

```python
# Q2- load the following image and apply sharpen filter.
# “write your observation”
img2 = plt.imread('blurred.png')
kernel = np.array([[-1,-1,-1],
                   [-1, 9,-1],
                   [-1,-1,-1]],
                   dtype= np.float32)
out = cv2.filter2D(img2, -1, kernel, borderType= cv2.BORDER_ISOLATED)

plt.figure(figsize=(10,20))
plt.subplot(121)
plt.imshow(img2)
plt.title('Original', size=30)
plt.subplot(122)
plt.imshow(out)
plt.title('Sharpenned', size=30)
plt.tight_layout()
```

```python
# Q2- load the following image and apply sharpen filter.
# “write your observation”
img3 = plt.imread('sharpen.png')
out2 = cv2.blur(img3, (20,20))

plt.figure(figsize=(10,20))
plt.subplot(121)
plt.imshow(img3)
plt.title('Original', size=30)
plt.subplot(122)
plt.imshow(out2)
plt.title('Blurred', size=30)
plt.tight_layout()
```
